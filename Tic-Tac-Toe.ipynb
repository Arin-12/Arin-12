```python
def summ(a,b,c):
	return a+b+c

def printfield(xstate, ostate):

	board=[('X' if xstate[i] else ('O' if ostate[i] else i )) for i in range(9)]

	print(f" {board[0]} | {board[1]} | {board[2]} ")
	print(f" ---|----|--- ")
	print(f" {board[3]} | {board[4]} | {board[5]} ")
	print(f" ---|----|--- ")
	print(f" {board[6]} | {board[7]} | {board[8]} ")


def checkwin(xstate ,ostate):
	wins=[[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]]
	for win in wins:
		if (summ(xstate[win[0]] , xstate[win[1]], xstate[win[2]])==3):
			print("Player 1 wins (X):")
			return 1
		if (summ(ostate[win[0]] , ostate[win[1]], ostate[win[2]])==3):
			print("Player 2 wins (O):")
			return 0

	return -1


if  __name__=="__main__" :
	xstate=[0,0,0,0,0,0,0,0,0]
	ostate=[0,0,0,0,0,0,0,0,0]
	turn=1
	moves=0
	print("\tWelcom to Tic Tac Toe Game")
	while (True):

		printfield(xstate,ostate)
		value=int(input(f"Player{turn} ({'X' if turn==1 else 'O'}) Enter position between (0-8):"))

		if value<0 or value>8 or xstate[value] or ostate[value]:
			print("Invalid Move ! Try again")
			continue
		if turn==1:
			xstate[value]=1
			turn=0
		else:
		 	ostate[value]=1
		 	turn=1

		moves+=1


		if checkwin(xstate,ostate)!=-1:
			printfield(xstate,ostate)
			break
		if moves==9:
			printfield(xstate,ostate)
			print("Its draw")
			break
```
